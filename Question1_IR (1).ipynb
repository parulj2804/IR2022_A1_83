```python
def preprocess_input(input):
  new_input = re.sub('[^a-zA-Z]', ' ', str(input))
  new_input = remove_numbers(new_input)
  new_input = remove_punctuation(new_input)
  new_input = remove_whitespace(new_input)
  new_input = text_lowercase(new_input)
  new_input = new_input.split()
  lemma = WordNetLemmatizer()
  all_stopwords = stopwords.words('english')
  all_stopwords.remove('not')
  new_input = [lemma.lemmatize(word) for word in new_input if not word in set(all_stopwords)]
  new_input = ' '.join(new_input)
  return new_input
  
```

```python
def text_lowercase(text):
    return text.lower()
 
```

```python
def remove_whitespace(text):
    return  " ".join(text.split())
 
```

```python
import string
def remove_punctuation(text):
    translator = str.maketrans('', '', string.punctuation)
    return text.translate(translator)
 
```

```python
def remove_numbers(text):
    result = re.sub(r'\d+', '', text)
    return result
```

```python
def NOT(dict, word, document_list):
  j = 0
  Not_list = []
  list1 = dict[word]
  for i in range(len(document_list)):
    if (list1[j][0] != i or j == len(list1)):
      Not_list.append((i, 0))
    else:
      j = j + 1
  return Not_list
```

```python
def OR(list1,list2):
  or_result = []
  comp = 0
  i = 0
  j = 0
  while (i != len(list1) and j != len(list2)):
      if (list1[i][0] < list2[j][0]):
        or_result.append(list1[i])
        i+=1
      elif (list1[i][0] > list2[j][0]):
        or_result.append(list2[j])
        j+=1
      else:
        or_result.append((list1[i][0], list1[i][1]+list2[j][1]))
        i += 1
        j += 1
      comp += 1
  while (j != len(list2)):
      or_result.append(list2[j])
      j += 1
  while (i != len(list1)):
      or_result.append(list1[i])
      i += 1
  return or_result,comp
```

```python
def AND(data1,data2):
  and_result = []
  comp = 0
  i = 0
  j = 0
  while (i != len(data1) and j != len(data2)):
      if (data1[i][0] > data2[j][0]):
        j+=1
        
      elif (data1[i][0] < data2[j][0]):
        i+=1

      else:
        and_result.append((data1[i][0], data1[i][1]+data2[j][1]))
        i += 1
        j += 1
        comp += 1
  return and_result,comp
```

```python
def merge(uni_dict, word1, word2, op, merge_list1=None, not_op=None):
  list1 = uni_dict[word1]
  list2 = []
  if (merge_list1 is not None and not_op is not None):
    list1 = merge_list1
    list2 = not_op
  elif (merge_list1 is not None):
    list2 = merge_list1
  elif (not_op is not None):
    list2 = not_op
  else:
    # print(word2)
    list2 = uni_dict[word2]
  comparisons = 0
  answer_list = []
  # print(list1, "\n", list2)
  if (op == 'OR'):
    answer_list, comparisons = OR(list1,list2)
  if (op == 'AND'):
    answer_list, comparisons = AND(list1,list2)


  return answer_list, comparisons
```

```python
def final(input, op, unigram_dict, doc_list):
  merge_list, comparisons_total = [], 0
  words = input.split(' ')
  i = 0
  initial_flag = 0
  while i != (len(op)):
    ops = op[i].split(' ')
    not_op = None
    operation_toPerform = ops[0]
    comparisons = 0
    if (len(ops) == 2 and ops[1] == 'NOT'):
      not_op = NOT(unigram_dict, words[i+1], doc_list)
    if not_op is None:
      if (initial_flag == 0):
        initial_flag = 1 
        merge_list, comparisons = merge(unigram_dict, words[i], words[i+1], operation_toPerform)
      else:
        merge_list, comparisons = merge(unigram_dict, words[i+1], words[i+1], operation_toPerform, merge_list)
    else:
      if (initial_flag == 0):
        merge_list, comparisons = merge(unigram_dict, words[i], words[i+1], operation_toPerform, not_op = not_op)
        initial_flag = 1
      else:
        merge_list, comparisons = merge(unigram_dict, words[i+1], words[i+1], operation_toPerform, merge_list, not_op = not_op)
    comparisons_total += comparisons
    i += 1
  print("Number of documents retrieved:", len(merge_list), "\nTotal comparisons done:", comparisons_total)
  doc_retrieved = []
  for docId in merge_list:
    try:
      doc_retrieved.append(doc_list[docId[0]])
    except:
      pass
  print("\nDocuments list:",doc_retrieved)
  print(merge_list)
  return merge_list, comparisons_total
```
